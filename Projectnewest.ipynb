```python
import numpy as np #importing numpy 
import pandas as pd #imported pandas 
import matplotlib.pyplot as plt#import matplotlib
real_GPD = pd.read_csv('CMSE201_data-3.csv')#data set I would use csv
real_GPD#seeing the dataset
```

# Results

```python
ThenewestGDPTexas=real_GPD.iloc[0,26:32]#from 2017 to 2022
ThenewestGDPTexas.mean()#only for the newestGDP
```

```python
time=[2017,2018,2019,2020,2021,2022]#time am using 
plt.figure(figsize=(10,12))
plt.subplot(3,2,1)#Created a subplot with 3 rows, 2 columns, and select the 1th subplot
plt.plot(time,Total_employmentTexas,color='orange')#I am plotting the Total_employment over the time
plt.title('Total employment In Texas before and after pandemic')#title
plt.xlabel('Years')#xaxis labeled year
plt.ylabel('Total_employment')#yaxis labeled Total_employment
plt.axvline(2019, color='red', linestyle='--', label='Start pandemic in 2019')#Added a vertical line at the year pandemic 2019 using a red dashed line
plt.legend()#used to put what's in my labeled

plt.subplot(3,2,2)#Created a subplot with 3 rows, 2 columns, and select the 2th subplot
plt.plot(time,ThenewestGDPTexas,color='blue')
plt.title('Texas Real GDP before and after pandemic')#title
plt.xlabel('Years')#xaxis labeled year
plt.ylabel('Real GDP')#ylabeled Real GDP
plt.axvline(2019, color='red', linestyle='--', label='Start pandemic in 2019')#Added a vertical line at the year pandemic 2019 using a red dashed line
plt.legend()#used to put what's in my labeled



plt.subplot(3,2,3)#Created a subplot with 3 rows, 2 columns, and select the 3th subplot
plt.plot(time,RealGDPAlabama,color='black')
plt.title('Alabama Real GDP before and after pandemic')#title
plt.xlabel('Years')#xaxis labeled year
plt.ylabel('Real GDP')#yaxis labeled Real GDP
plt.axvline(2019, color='red', linestyle='--', label='Start pandemic in 2019')#Added a vertical line at the year pandemic 2019 using a red dashed line
plt.legend()#used to put what's in my labeled

plt.subplot(3,2,4)#Created a subplot with 3 rows, 2 columns, and select the 4th subplot
plt.plot(time,Total_employmentAlabama,color='blue')
plt.title('Total employment In Alabama before and after pandemic')#title
plt.xlabel('Years')#xlabeled year
plt.ylabel('Total employment')#yaxis labeled Total_employment
plt.axvline(2019, color='red', linestyle='--', label='Start pandemic in 2019')#Added a vertical line at the year pandemic 2019 using a red dashed line
plt.legend()#used to put what's in my labeled
plt.grid()#better readability
plt.tight_layout()# Adjusting the layout to prevent clipping of titles and labels
plt.show()# Display the plot

```

```python
plt.figure(figsize=(10,12))
plt.subplot(3,2,1)#Created a subplot with 3 rows, 2 columns, and select the 1th subplot
plt.plot(time,ThenewestGDP_US,color='black')#plotting REAL GDP over the time with black color in the U.S.A
plt.title('U.S.A Real GDP before and after pandemic')#title
plt.xlabel('Year')#xaxis labeled year
plt.ylabel('Real GDP')#yaxis labeled Real GDP
plt.axvline(2019, color='blue', linestyle='--', label='Start pandemic in 2019')#Added a vertical line at the year pandemic 2019 using a blue dashed line
plt.legend()#used to put what's in my labeled

plt.subplot(3,2,2)#Created a subplot with 3 rows, 2 columns, and select the 2th subplot
plt.plot(time,Total_employmentUS,color='black')#plotting Total employment over the time with black color in the U.S.A
plt.title('Total employment In U.S.A before and after pandemic')#title
plt.xlabel('Year')#xaxis labeled year
plt.ylabel('Total_employment ')#yaxis labeled Total_employment
plt.axvline(2019, color='black', linestyle='--', label='Start pandemic in 2019')#Added a vertical line at the year pandemic 2019 using a black dashed line
plt.legend()#used to put what's in my labeled

plt.subplot(3,2,3)#Created a subplot with 3 rows, 2 columns, and select the 3th subplot
plt.plot(time,RealGDPAlaska,color='black')
plt.title('Alaska Real GDP before and after pandemic')#title
plt.xlabel('Year')#axis labeled year
plt.ylabel('Real GDP ')#yaxis labeled Real GDP
plt.axvline(2019, color='lime', linestyle='--', label='Start pandemic in 2019')#Added a vertical line at the year pandemic 2019 using a lime dashed line
plt.legend()#used to put what's in my labeled

plt.subplot(3,2,4)#Created a subplot with 3 rows, 2 columns, and select the 4th subplot
plt.plot(time,Total_employmentAlaska,color='blue')#plot of Total employment in Alaska over the time with color blue.
plt.title('Total employment In Alaska before and after pandemic')#title
plt.xlabel('Year')#xaxis labeled year
plt.ylabel('Total employment')#yaxis labeled Total_employment
plt.axvline(2019, color='red', linestyle='--', label='Start pandemic in 2019')#Added a vertical line at the year pandemic 2019 using a red dashed line
plt.legend()#used to put what's in my labeled
plt.grid()#better readability
plt.tight_layout()# Adjusting the layout to prevent clipping of titles and labels
plt.show()# Display the plot
```
